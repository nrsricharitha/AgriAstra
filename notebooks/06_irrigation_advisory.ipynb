```python
import pandas as pd
```

```python
df = pd.read_csv('/content/agriastra_ai_dataset.csv')

df.head()
```

```python
df.columns
```

```python
def irrigation_advice(stress):

    if stress == 'Low':
        return 'No irrigation needed'

    elif stress == 'Moderate':
        return 'Irrigate within 3 days'

    else:
        return 'Irrigate immediately'
```

```python
df['Irrigation_Advisory'] = df['Moisture_Stress'].apply(

    irrigation_advice

)

df.head()
```

```python
df['Irrigation_Advisory'].value_counts()
```

```python
df.to_csv(

    'agriastra_final_dataset.csv',

    index=False

)

print("Final dataset created.")
```
